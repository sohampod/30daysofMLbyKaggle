```python
def gini(smaples):
    samples_sum = sum(samples)
    score = 0
    probas = []
    for sample in samples:
        proba = sample/samples_sum
        probas.append(proba)
    for proba in probas:
        score += proba * (1-proba)
    return score
```

```python
samples =[60,120,50]
gini(samples)
```
